```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
import plotly.colors as colors
pio.templates.default = "plotly_white"
```

```python
data = pd.read_csv("Sample - Superstore.csv",encoding = 'latin-1')
```

```python
data.head(5)
```

```python
data.describe()
```

```python
data.info()
```

Converting date columns

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data.info()
```

```python
data.head(5)
```

```python
data['Order Month'] = data['Order Date'].dt.month
data['Order Year'] = data['Order Date'].dt.year
data['Order Day of week'] = data['Order Date'].dt.dayofweek
```

```python
data.head(5)
```

# Monthly sales Analysis

```python
sales_by_month = data.groupby('Order Month')['Sales'].sum().reset_index()
```

```python
sales_by_month
```

```python
fig = px.line(sales_by_month,
              x = 'Order Month',
              y = 'Sales',
              title = 'Monthly sales Analysis')
fig.show()
```

```python
data.head(5)
```

```python
sales_by_category = data.groupby('Category')['Sales'].sum().reset_index()
```

```python
sales_by_category
```

```python
fig = px.pie(sales_by_category,
             values = 'Sales', # Changed 'sales' to 'Sales'
             names = 'Category', # Changed 'category' to 'Category'
             hole = 0.4,
             color_discrete_sequence= px.colors.qualitative.Pastel)
fig.update_traces(textposition = 'inside', textinfo = 'percent+label')
fig.update_layout(title_text = 'sales Analysis by category', title_font= dict(size = 25)) # Changed 'titlr_font' to 'title_font'
fig.show()
```

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
```

```python
color_palatte = colors.qualitative.Pastel
fig = go.Figure()
```

```python
fig.add_trace(go.Bar(x= sales_profit_by_segment['Segment'],
                     y =sales_profit_by_segment['Sales'],
                     name ='Sales',
                     marker_color = color_palatte[0]))
fig.add_trace(go.Bar(x= sales_profit_by_segment['Segment'],
                     y =sales_profit_by_segment['Profit'],
                     name ='Profit',
                     marker_color = color_palatte[1]))
fig.update_layout(title = 'Sales and Profit Analysis by Customer Segment',
                  xaxis_title= 'Customer Segment',yaxis_title='Amount')
fig.show()
```

#Sales to Profit Ratio

```python
sales_profit_by_segment = data.groupby('Segment').agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
```

```python
sales_profit_by_segment['sales_profit_Ratio'] = sales_profit_by_segment['Sales'] / sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment', 'sales_profit_Ratio']])
```
